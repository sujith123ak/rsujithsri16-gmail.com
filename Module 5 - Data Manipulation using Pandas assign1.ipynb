```python
import pandas as pd

# Load the CSV file into a DataFrame
churn = pd.read_csv('customer_churn.csv')
```

```python
# Select the 3rd, 7th, 9th, and 20th columns (indexing starts from 0)
newCols = churn.iloc[:, [2, 6, 8, 19]]
print(newCols.head())
```

```python
# Select rows from index 200 to 1000 (inclusive)
selected_rows = churn.iloc[200:1001]
print(selected_rows.head())
```

```python
# Select rows from index 20 to 200 (exclusive) and columns from index 2 to 15
subset = churn.iloc[20:200, 2:15]
print(subset.head())
```

```python
# Display the top 100 records
top_100_records = churn.head(100)
print(top_100_records)
```

```python
# Display the last 10 records
last_10_records = churn.tail(10)
print(last_10_records)
```

```python
# Display the last record
last_record = churn.tail(1)
print(last_record)
```

```python
# Sort by the 'tenure' column in descending order
sorted_by_tenure = churn.sort_values(by='tenure', ascending=False)
print(sorted_by_tenure.head())
```

```python
# a. Tenure > 50 and gender as 'Female'
condition_a = churn[(churn['tenure'] > 50) & (churn['gender'] == 'Female')]
print(condition_a)
```

```python
# b. Gender as 'Male' and SeniorCitizen as 0
condition_b = churn[(churn['gender'] == 'Male') & (churn['SeniorCitizen'] == 0)]
print(condition_b)
```

```python
# c. TechSupport as 'Yes' and Churn as 'No'
condition_c = churn[(churn['TechSupport'] == 'Yes') & (churn['Churn'] == 'No')]
print(condition_c)
```

```python
# d. Contract type as 'Month-to-month' and Churn as 'Yes'
condition_d = churn[(churn['Contract'] == 'Month-to-month') & (churn['Churn'] == 'Yes')]
print(condition_d)
```

```python
# Use a for loop to count male senior citizens receiving tech support
count = 0
for index, row in churn.iterrows():
    if row['TechSupport'] == 'Yes' and row['gender'] == 'Male' and row['SeniorCitizen'] == 1:
        count += 1

print("Number of male senior citizens receiving tech support:", count)
```
