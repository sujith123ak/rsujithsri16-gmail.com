```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the dataset
df = pd.read_csv('mtcars.csv')
```

```python
# Step 2: Group the data by 'carbs' and 'gear' and count the occurrences
grouped_data = df.groupby(['carb', 'gear']).size().unstack(fill_value=0)
```

```python
# Step 3: Plot the bar graph
grouped_data.plot(kind='bar', stacked=False)
```

```python
# Step 4: Customize the plot
plt.xlabel('Number of carburetors')  # x-axis label
```

```python
plt.ylabel('Number of forward gears')  # y-axis label
```

```python
plt.title('carb vs gear')  # Title of the plot
```

```python
# Step 5: Display the plot
plt.show()
```
