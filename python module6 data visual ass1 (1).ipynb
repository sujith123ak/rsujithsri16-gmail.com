```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
df = pd.read_csv('cars.csv')
```

```python
# Step 2: Plot the line graph
plt.figure(figsize=(10, 6))  # Optional: Set the figure size for better visibility
plt.plot(df['model'], df['hp'], marker='o')  # Plot the data with a line and markers
```

```python
# Step 3: Customize the plot
plt.xlabel('Models of the cars')  # x-axis label
```

```python
plt.ylabel('Horse-Power of Cars')  # y-axis label
```

```python
plt.title('Model Names vs HorsePower')  # Title of the plot
```

```python
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
```

```python
# Step 4: Display the plot
plt.show()
```
