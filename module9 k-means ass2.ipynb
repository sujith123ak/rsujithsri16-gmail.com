```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Step 1: Load the customer data
data = pd.read_csv('customer_data.csv')
```

```python
# Step 2: Check the number of cells in each column with null values
null_values = data.isnull().sum()
print("Null values in each column:\n", null_values)
```

```python
# Step 3: Create a scatter plot with Age as X and Spending Score as Y
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(data['Age'], data['Spending Score (1-100)'], c='blue')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Spending Score (1-100)')
```

```python
plt.title('Scatter Plot of Age vs Spending Score')
```

```python
plt.show()
```

```python
# Step 4: Use the Elbow Method to find the best number of clusters
X = data[['Age', 'Spending Score (1-100)']]
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the Elbow Method
plt.figure(figsize=(8, 6))
```

```python
plt.plot(range(1, 11), wcss, marker='o')
```

```python
plt.title('Elbow Method')
```

```python
plt.xlabel('Number of Clusters')
```

```python
plt.ylabel('WCSS')
```

```python
plt.show()
```

```python
optimal_clusters = 3  # Replace with the optimal number from the elbow plot
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

```python
# Scatter plot with clusters
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(data['Age'], data['Spending Score (1-100)'], c=data['Cluster'], cmap='viridis')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Spending Score (1-100)')
```

```python
plt.title('Scatter Plot of Age vs Spending Score with Clusters')
```
