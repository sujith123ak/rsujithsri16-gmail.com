```python
import pandas as pd

# Load the DataFrame from the CSV file
churn = pd.read_csv('customer_churn.csv')
```

```python
# 1. Sort by the 'tenure' column in descending order
sorted_churn = churn.sort_values(by='tenure', ascending=False)
print("Sorted DataFrame by tenure (descending):")
print(sorted_churn.head())
```

```python
# 2. Fetch records based on conditions
# a. Tenure > 50 and gender as 'Female'
condition_a = churn[(churn['tenure'] > 50) & (churn['gender'] == 'Female')]
print("\nRecords where Tenure > 50 and Gender is Female:")
print(condition_a)
```

```python
# b. Gender as 'Male' and SeniorCitizen as 0
condition_b = churn[(churn['gender'] == 'Male') & (churn['SeniorCitizen'] == 0)]
print("\nRecords where Gender is Male and SeniorCitizen is 0:")
print(condition_b)
```

```python
# c. TechSupport as 'Yes' and Churn as 'No'
condition_c = churn[(churn['TechSupport'] == 'Yes') & (churn['Churn'] == 'No')]
print("\nRecords where TechSupport is Yes and Churn is No:")
print(condition_c)
```

```python
# d. Contract type as 'Month-to-month' and Churn as 'Yes'
condition_d = churn[(churn['Contract'] == 'Month-to-month') & (churn['Churn'] == 'Yes')]
print("\nRecords where Contract type is Month-to-month and Churn is Yes:")
print(condition_d)
```
