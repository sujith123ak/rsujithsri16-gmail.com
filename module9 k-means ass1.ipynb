```python
import pandas as pd

# Load the dataset
data = pd.read_csv('customer_data.csv')
```

```python
# Check for null values
null_values = data.isnull().sum()
print(null_values)
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Age vs Spending Score
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(data['Age'], data['Spending Score (1-100)'], c='blue')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Spending Score (1-100)')
```

```python
plt.title('Scatter Plot of Age vs Spending Score')
```

```python
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
# Select features for clustering
X = data[['Age', 'Spending Score (1-100)']]
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42) 
data['Cluster'] = kmeans.fit_predict(X)
```

```python
# Scatter plot with clusters
plt.figure(figsize=(8, 6))
```

```python
plt.scatter(data['Age'], data['Spending Score (1-100)'], c=data['Cluster'], cmap='viridis')
```

```python
plt.xlabel('Age')
```

```python
plt.ylabel('Spending Score (1-100)')
```

```python
plt.title('Scatter Plot of Age vs Spending Score with Clusters')
```

```python
plt.show()
```
