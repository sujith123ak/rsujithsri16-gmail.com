```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
df = pd.read_csv('mtcars.csv')
```

```python
carb_data = df.groupby('am')['carb'].sum()
```

```python
# Step 3: Plot the area chart
plt.figure(figsize=(10, 6))  # Optional: Set the figure size for better visibility
plt.fill_between(carb_data.index, carb_data, color="skyblue", alpha=0.4)
```

```python
# Step 4: Customize the plot
plt.xlabel('Transmission')  # x-axis label
```

```python
plt.ylabel('Number of carburetors')  # y-axis label
```

```python
plt.title('Transmission vs Number of carburetors')  # Title of the plot
```

```python
plt.xticks(ticks=[0, 1], labels=['Automatic', 'Manual'])  # Set custom labels for 'am'
```

```python
# Step 5: Display the plot
plt.show()
```
